```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('~/Downloads/train.csv')
data.head(1)
```

```python
#Membagi variabel Numerik dan Categori
#TARGET VARIABEL SURVIVED

numeric_data = data.select_dtypes(include=[np.number])
categorical_data = data.select_dtypes(exclude=[np.number])
numeric_data
# categorical_data
# numeric_data.Age.mean
```

```python
# Pembuatan barplot (horizontal dan vertikal)
# Pembuatan distplot
# Pembuatan boxplot
# Pembuatan scatterplot
# Pembuatan pie chart
```

```python
#Vertical barplot
#variabel sex dan survived
sns.barplot(x=categorical_data['Sex'],y=numeric_data['Survived'],hue=categorical_data['Embarked'],data=data).set(title='Jumlah yang selamat berdasarkan embarked')
sns.set(rc={'figure.figsize':(5,3)})
# plt.savefig('oo.png')
```

```python
#Horizontal barplot
#variabel sex dan survived 
sns.set_theme(style="whitegrid")
sns.barplot(data=data,y=categorical_data['Sex'],x=numeric_data['Survived'],orient='h').set(title='Jumlah yang selamat')
```

```python
# distribution plot
#Dist Fare 
#menentukan jumlah selamat atau tidak nya berdasar umur

sns.set(style="whitegrid", palette="muted", color_codes=True)
fig = plt.figure(constrained_layout=True, figsize=(8, 3))
spec = fig.add_gridspec(1, 2)

ax = fig.add_subplot(spec[0, :])
sns.histplot(data=data, x=numeric_data['Age'], bins=np.arange(-0.5, 86, 5), kde=False, hue=numeric_data['Survived'], legend=True, ax=ax).set(title='Jumlah selamat berdasar Umur')
ax.legend(['Yes', 'No'], title='Survived?', prop={'size': 12})
plt.show()
```

```python
#boxplot
# variabel Survived dan Fare
sns.boxplot(data=data,x=numeric_data['Survived'],y=numeric_data['Fare']) 
sns.set(rc={'figure.figsize':(10,5)})
plt.title('Jumlah survived berdasar Fare')
plt.xticks(rotation=0)
plt.show()
```

```python
#Scatterplot
#variabel fare dan Survived
sns.set(rc={'figure.figsize':(10,5)})
sns.set_theme(style="darkgrid")#set grid putih
sns.scatterplot(data=data,x=numeric_data['Fare'],y=numeric_data['Survived'],hue=numeric_data['Pclass'],palette='pastel').set(title='fare berdasar survived')
plt.show()
```

```python
#PIE chart
#persentase jumlah selamat
#variabel survive dan sex

jumlahlakidanwanita=data.groupby('Sex').sum()['Survived'].reset_index()
jumlahlakidanwanita


f,a = plt.subplots()
plt.title('Persentase Jumlah yang selamat')
a.pie(jumlahlakidanwanita['Survived'],labels=jumlahlakidanwanita['Sex'],autopct='%1.1f%%',startangle=90)
plt.show()
```
