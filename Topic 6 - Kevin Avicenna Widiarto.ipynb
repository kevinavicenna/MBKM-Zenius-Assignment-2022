# Join

Using two of the dataframes below, answer the questions :

```python
#import libraries here
import pandas as pd
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

```python
#code goes here
df3 = pd.merge(df1,df2,on='key')
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

```python
df4 = pd.merge(df1,df2,on='key',how='left')
df4
```

#### 1. Replace key "K2" into "K3" on df1
#### 2. Left join it with df2 
(15pts)

```python
#code goes here
df_rep = df1.replace({'key':{'K2':'K3'}})
left_join_df1_df2_replace = pd.merge(df_rep,df2,how = 'left') 
left_join_df1_df2_replace
```

## Using data aggregation in pandas , Answer the question below:

Show the total amount of passengers who survived and not survived (10pts)

```python
#code goes here
SURVIVED = df.Survived.sum() #variabel survived menjumlah nilai kolom survived sbg yang hidup
TOTAL = df.Survived.count() #variabel Total menghitung row 
NOT_SURVIVED = TOTAL - SURVIVED #variabel not survived ,total keseluruhan row dikurangi jumlah nilai survived menghasilkan nilai orang yg tidak survived
print("Yang hidup totalnya",SURVIVED)
print("Yang meninggal totalnya",NOT_SURVIVED)
```

Show the average fare paid based on gender (10 pts)

```python
#code goes here

#MALE
male = df.loc[df['Sex'] == "male"] #var male menselect kolom sex bernilai male
res_male_fare = male.Fare.sum() / male.Fare.count() #var res_male_fare , sum nilai kolom Fare dibagi jumlah count row Fare

#FEMALE
female = df.loc[df['Sex'] == "female"] #var female menselect kolom sex bernilai female
res_female_fare = female.Fare.sum() / female.Fare.count() #var res_female_fare , sum nilai kolom Fare dibagi jumlah count row Fare

print("Fare Male :", res_male_fare)
print("Fare Female :",res_female_fare)
```

Show the total amount of passengers who survived and not survived based on gender (10pts)

```python
#code goes here
male = df.loc[df['Sex'] == "male"]#var male menselect kolom sex bernilai male

male_survived = male.loc[df['Survived']==1] #var male menselect kolom survived bernilai 1
male_not_survived = male.loc[df['Survived']==0] #var male menselect kolom survived bernilai 0

print("Laki-Laki \nTotal Jumlah Laki yang selamat ",len(male_survived))
print("Total Jumlah Laki yang tidak selamat ",len(male_not_survived))



female = df.loc[df['Sex'] == "female"] #var female menselect kolom sex bernilai male

female_survived = female.loc[df['Survived']==1] #var female menselect kolom survived bernilai 1
female_not_survived = female.loc[df['Survived']==0] #var female menselect kolom survived bernilai 0

print("\nWanita \nTotal Jumlah Wanita yang selamat ",len(female_survived))
print("Total Jumlah Wanita yang tidak selamat ",len(female_not_survived))
```
