### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

```python
#Dist Variable Age
fig,ax = plt.subplots(figsize=(7,4))
sns.set_style('whitegrid')
sns.histplot(df['AGE'],color='00000',bins=30,kde=False).set(title='Age Variable distribution')
```

```python
#Dist variable Sex
fig,ax = plt.subplots(figsize=(2,4))
sns.set_style('whitegrid')
sns.distplot(df['SEX'],color='red',bins=30,kde=False).set(title='Sex Variable distribution')
```

```python
#Dist Variabel BMI
fig,ax = plt.subplots(figsize=(7,4))
sns.set_style('whitegrid')
sns.histplot(df['BMI'],color='magenta',bins=30,kde=False).set(title='BMI Variable distribution')

```

```python
#Dist variabel Y
fig,ax = plt.subplots(figsize=(7,4))
sns.set_style('whitegrid')
sns.histplot(df['AGE'],color='green',bins=30,kde=False).set(title='Y Variable distribution')

#Merupakan normal distribution
```

# SKEWNESS
menurut https://medium.com/@atanudan/kurtosis-skew-function-in-pandas-aa63d72e20de

 Jika skewness antara -0,5 dan 0,5, datanya cukup simetris

 Jika kemiringan antara -1 dan  0,5 atau antara 0,5 dan 1, data agak miring

 Jika kemiringan kurang dari -1 atau lebih besar dari 1, data sangat miring







# Kesimpulan

Age (skewness negatif) = termasuk simetris ,ekor lebih miring ke kiri

Sex (skewness positif) = termasuk simetris, ekor lebih miring ke kanan mendekati nilai normal distribusi

BMI (skewness positif) = termasuk miring, ekor miring besar disebelah kanan

Y   (skewness positif) = termasuk simetris,ekor lebih miring ke kiri
